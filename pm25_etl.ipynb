```python
import pandas as pd
import glob
```

```python
df1 = pd.read_csv("./Data/PM25/Particulas-PM2.5-data-2022.csv")
df2 = pd.read_csv("./Data/PM25/Particulas-PM2.5-data-2023.csv")
df3 = pd.read_csv("./Data/PM25/Particulas-PM2.5-data-2024.csv")
df4 = pd.read_csv("./Data/PM25/Particulas-PM2.5-data-2025.csv")
```

```python
datos = pd.concat(
    [df1, df2, df3,df4],
    ignore_index=True
)

datos.head()
```

```python
datos.dtypes
```

```python
datos["Particulas_PM25"] = pd.to_numeric(
    datos["Particulas_PM25"]
    .str.replace(" μg/m³", "", regex=False),
    errors="coerce"
)
```

```python

datos["Time"] = pd.to_datetime(datos["Time"])
```

```python
datos["YEAR"] = datos["Time"].dt.year
datos["MONTH"] = datos["Time"].dt.month
datos["DAY"] = datos["Time"].dt.day
```

```python
pm25_diario = (
    datos
    .groupby(["YEAR","MONTH","DAY"])
    .agg(
        PM25=("Particulas_PM25","mean")
        #MEDICIONES=("Particulas_PM25","count")
    )
    .reset_index()
)
pm25_diario["PM25"] = pm25_diario["PM25"].round(1)
```

```python
pm25_diario.to_csv(
    "./Data/PM25/PM25_DAIRY_FULL_2022_2025.csv",
    index=False
)
```

```python
pm25_diario.shape
```

```python



# 2. Agrupar por AÑO y MES, y calcular la media de PM25
# Usamos .reset_index() para que YEAR y MONTH vuelvan a ser columnas normales
df_mensual = pm25_diario.groupby(['YEAR', 'MONTH'])['PM25'].mean().reset_index()

# 3. Renombrar la columna para reflejar que ahora es un promedio
df_mensual = df_mensual.rename(columns={'PM25': 'MEDIA_PM25'})

# 4. Mostrar el resultado en consola
print(f"Número de filas del nuevo dataset: {len(df_mensual)}")
print(df_mensual.head())

# 5. Opcional: Guardar el nuevo dataset en un archivo CSV limpio
df_mensual.to_csv('pm25_media_mensual.csv', index=False)
```
