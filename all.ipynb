```python
import csv
import pandas as pd
```

```python
# CONFIGURACIÓN
INPUT_TXT = "./Data/Vehiculos/PARQUE-VEHICULAR-2007-2025-FULL.txt"
OUTPUT_CSV = "./Data/Vehiculos/PARQUE-VEHICULAR-2007-2025-FULL.csv"
ENCODING_INPUT = "cp1252"
ENCODING_OUTPUT = "utf-8"
```

```python
print("=" * 80)
print("PASO 1: Leyendo archivo txt...")
print("=" * 80)

rows = []
with open(INPUT_TXT, "r", encoding=ENCODING_INPUT, newline="") as infile:
    reader = csv.reader(infile, delimiter="|", quotechar='"', escapechar="\\")
    for row in reader:
        rows.append(row)

if not rows:
    raise ValueError("El archivo de entrada está vacío.")

print(f"Filas leídas: {len(rows):,}")
```

```python
header = rows[0]
expected_columns = len(header)

filas_validas = [header]
filas_invalidas = []

for line_number, row in enumerate(rows[1:], start=2):
    if len(row) != expected_columns:
        filas_invalidas.append((line_number, row))
    else:
        filas_validas.append(row)

print("Validas:", len(filas_validas))
print("Invalidas:", len(filas_invalidas))
```

```python
df = pd.DataFrame(filas_validas[1:], columns=filas_validas[0])
df.columns = df.columns.str.strip()
df["ANIO_ALZA"] = pd.to_numeric(df["ANIO_ALZA"], errors="coerce")
df.head()
```

```python
df_filtrado = df[
    (df["NOMBRE_DEPARTAMENTO"] == "GUATEMALA") &
    (df["NOMBRE_MUNICIPIO"] == "GUATEMALA") &
    (df["ANIO_ALZA"] < 2025)
]

df_filtrado.shape
```

```python
tipo_vehiculo = sorted(df_filtrado["TIPO_VEHICULO"].dropna().unique())
tipo_vehiculo
```

```python
livianos = {
    'AUTOMOVIL','AMBULANCIA','JEEP','PICK UP','CARRO PARA GOLF',
    'CAMIONETA','CAMIONETA AGRICOLA','CAMIONETA SPORT','CAMIONETILLA',
    'PANEL','MICROBUS','MICROBUS ESCOLAR','MINIBUS','MOTO','SCOOTER',
    'TRIMOTO','CUATRIMOTO','LIMOSINA','CARRO FUNEBRE','VEHICULO RUSTICO'
}
```

```python
df_livianos = df_filtrado[df_filtrado["TIPO_VEHICULO"].isin(livianos)]
df_livianos.shape
```

```python
df_livianos.to_csv(OUTPUT_CSV, index=False, encoding=ENCODING_OUTPUT)
print("Archivo guardado:", OUTPUT_CSV)
filas, columnas = df_livianos.shape

print("Filas:", filas)
print("Columnas:", columnas)
```
